```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.DataFrame(pd.read_excel("View_My_Courses (4).xlsx"))
df.head()
```

```python
df2 = df.drop(0, axis='index')
df2.head()
```

```python
nameDetails = df2["My Enrolled Courses"]
nameDetails.drop(1, axis="index", inplace=True)
df2.drop("My Enrolled Courses", axis='columns', inplace=True)
nameDetails.head()
```

```python
# split the name information
details_split = nameDetails.str.rsplit(" - ", expand=True, n=1)
name_id_split = details_split[0].str.split(" - ", expand=True, n=1)[0]
terms = details_split[1].str.rsplit(" (", expand=True, n=1)[0]
name_id = name_id_split.iloc[0]
terms
# print(name_id)
```

```python
# drop first column
colname = list(df2.columns)
colname[0]
df3 = df2.drop(colname[0], axis='columns')
# df3 = df2.drop(0, axis='columns')
colname = list(df3.loc[1])
df3.columns = colname
df3.drop(1, axis="index", inplace=True)
df3['Term'] = terms
df3.drop(["Drop", "Swap", "Credits", "Grading Basis"], axis="columns", inplace=True)
df3.head()

```

```python
# splitting the meeting pattern
# df4 = df3
mpa_raw = df3["Meeting Patterns"]
mpa_raw[2]
mpa_split = mpa_raw.str.split("\n\n", expand=True)
mpa_nodate = mpa_split[0].str.split("|", expand=True, n=1)[1]
mpa_nodate1 = mpa_split[1].str.split("|", expand=True, n=1)[1]
mpa_split1 = pd.concat([mpa_nodate, mpa_nodate1], axis=1)
mpa_split1.columns = [0, 1]

mpa_single = mpa_split1[mpa_split1[0] == mpa_split1[1]][0]
mpa_multiple = mpa_split1[(mpa_split1[0] != mpa_split1[1]) & (mpa_split1[1].notna())]
mpa_multiple
```

```python
# adding meeting pattern to df, supports course with two meeting patterns
# reset index
df4 = df3.drop(["Meeting Patterns"], axis="columns")
df4["mpa0"] = mpa_nodate
df4["mpa1"] = mpa_nodate1
df4.reset_index(inplace=True)
mpa_multiple_df = df4[(df4["mpa0"] != df4["mpa1"]) & (df4["mpa1"].notna())]


for row in mpa_multiple_df.index:
    ind = len(df4)
    df4.loc[ind] = df4.loc[row].copy()
    df4.loc[ind, "mpa0"] = df4.loc[row, "mpa1"]

df4.drop("mpa1", axis="columns", inplace=True)
df4
```

```python
# Get buidling code, floor, room
df5 = df4
mpa_parse = df4['mpa0'].str.split(" \| ", expand=True)
buildings = mpa_parse[3].str.split(" \(", expand=True)[1].str.split("\)", expand=True)[0]

df5["Building"] = buildings


# Get days
df5["Days"] = mpa_parse[0]

floor = mpa_parse[4].str.split(":", expand=True)
room = mpa_parse[5].str.split(":", expand=True)
mpa_parse["Room"] = floor[0] + floor[1] + "-" + room[0] + room[1]

df5["Room"] = mpa_parse["Room"]

mpa_parse
```

```python
from gclss_prep import get_time

time = mpa_parse[1]
# mpa[2].str.split(" \- ", expand=True)
df5["Start"] = get_time("Start", time)
df5["End"] = get_time("End", time)
df5.drop("mpa0", axis="columns", inplace=True)
df5
```

```python

course = gpd.GeoDataFrame(df5)
bcen = gpd.read_file("geo_files/ubc_buildings_centroids.geojson")

bcen1 = bcen[['BLDG_CODE', 'NAME', 'SHORTNAME', 'POSTAL_CODE', 'PRIMARY_ADDRESS', 'geometry']]

gclss = course.merge(bcen1, left_on='Building', right_on='BLDG_CODE')
gclss.drop("BLDG_CODE", axis='columns', inplace=True)
gclss1 = gpd.GeoDataFrame(gclss)

gclss1
```
